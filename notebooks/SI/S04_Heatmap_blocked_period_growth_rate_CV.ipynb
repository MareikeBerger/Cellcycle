```python
data_frame = makeDataframe.make_dataframe(file_path)
data_frame = makeDataframe.add_average_values_to_df(data_frame)
data_frame
```

# Fraction of initiated origins

```python
def return_array_of_initiation_volumes(filepath_h5):
    data_frame = pd.read_hdf(filepath_h5, key='dataset_init_events')
    return data_frame['v_init_per_ori'], data_frame['t_init']

def calculate_CV_v_init(row):
    v_init_per_ori, t_init = return_array_of_initiation_volumes(row.path_dataset)
    v_init_np = v_init_per_ori.to_numpy()
    return np.std(v_init_np[10:])/np.mean(v_init_np[10:])

def calculate_v_init_mean(row):
    v_init_per_ori, t_init = return_array_of_initiation_volumes(row.path_dataset)
    v_init_np = v_init_per_ori.to_numpy()
    return np.mean(v_init_np[10:])

data_frame["CV_v_init"] = data_frame.apply(lambda row: calculate_CV_v_init(row), axis=1)
data_frame["v_init_mean"] = data_frame.apply(lambda row: calculate_v_init_mean(row), axis=1)
data_frame.CV_v_init
```

# Mark aborted simulations in dataframe

```python
def determine_whether_simulation_aborted(filepath_h5):
    data_frame_traces = pd.read_hdf(filepath_h5, key='dataset_time_traces')
    return data_frame_traces["abort_simulation"][0]

data_frame.loc[:, 'aborted'] = data_frame.apply(lambda row: determine_whether_simulation_aborted(row.path_dataset), axis = 1)
data_frame.loc[:, 'rate_growth_rounded'] = data_frame.apply(lambda row: np.round(row.rate_growth, 2), axis = 1)
data_frame.loc[:, 'period_blocked_min'] = data_frame.apply(lambda row: np.round(row.period_blocked*60, 1), axis = 1)
```

# Plot CV

```python
import matplotlib.colors as colors
import matplotlib as mpl

data_frame_ordered = data_frame.sort_values(by=["rate_growth"])
heatmap_df = data_frame_ordered.pivot( "period_blocked_min","rate_growth_rounded", "CV_v_init")
heatmap_mask = data_frame_ordered.pivot( "period_blocked_min","rate_growth_rounded", "aborted")

# divnorm = colors.TwoSlopeNorm(vmin=data_frame["degree_synch"].min(), vcenter=1, vmax=2)
cmap = mpl.cm.get_cmap("plasma").copy()
# cmap.set_over('darkred')
sns.set(rc={'figure.figsize': (18.7, 3.27)})
ax = sns.heatmap(heatmap_df,
            mask=heatmap_mask,  
                 
            square=True, 
            linewidths=.5,
            cmap=cmap,
#             vmin=0, 
#             vmax=2, 
#             center=1.00,
#             cbar_kws={'extend': 'max'}
                )
ax.invert_yaxis()
# ax.vlines(x=0.5, ymin=0, ymax=0.36)

# ax.set_title(r'$s= \frac{n_{\rm ori}(t_{\rm i})}{n_{\rm ori}(t_{\rm f})- n_{\rm ori}(t_{\rm i})}$')
ax.set_facecolor('grey')
ax.set(xlabel=r'$\lambda$ [h$^{-1}$]', ylabel=r'$\tau_b$ [min]')
plt.savefig(file_path + '/heatmap_CV.pdf', format='pdf', bbox_inches='tight')
```
